```python
#Q9. How do you concatenate two Pandas DataFrames?
'''Let me take 2 dataframes and concatenate both'''
data ={'name':['shanthi','karur','krishna','lakshmi'], 'age':[22,21,24,54]}
data2 ={'name':['ankita','aruna','shanthi','karur'], 'age':[20,24,44,30]}
df=pd.DataFrame(data)
df1=pd.DataFrame(data2)
ser=pd.concat([df,df1])
ser
```

```python
#Q10. How do you merge two Pandas DataFrames on a specific column?
pd.merge(df,df1,on ='name')
```

```python
#Q11. How do you group data in a Pandas DataFrame by a specific column and apply an aggregation function?
ser.groupby('name')['age'].sum()
```
